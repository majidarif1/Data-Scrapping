```python
all_data = []
page=1
while True:  
    url = f'https://jomo.pk/collections/men-footwear?page={page}'
    session = requests.Session()
    response = session.get(url = url)
    yml = '''
    Product_data:
            css: 'li.grid__item'
            xpath: null
            multiple: true
            type: Text
            children:
                name:
                    css: 'div.h4'
                    xpath: null
                    multiple: true
                    type: Text
                category:
                    css: 'div.price__vendor'
                    xpath: null
                    multiple: true
                    type: Text
                price:
                    css: 'div.price__regular span.price-item'
                    xpath: null
                    multiple: true
                    type: Text
                new_price:
                    css: 'div.price__sale span.price-item'
                    xpath: null
                    type: Text
    

    '''
    e = Extractor.from_yaml_string(yml)
    data = e.extract(response.text)
    print(url,"\n",data['Product_data'])
    page += 1
    try:
        all_data += data['Product_data']
    except:
        break
```
