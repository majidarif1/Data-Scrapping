```python
from selectorlib import Extractor
import requests
import pandas as pd
import pymysql
```

```python
all_data = []
page=1
#while True:
url = f'https://www.rekotrade.com/marketplace_items_10_page_page={page}'
session = requests.Session()
response = session.get(url = url)
yml = '''
cars_details:
        css: 'div.thumbnail '
        xpath: null
        multiple: true
        type: Text
        children:
                
            mileage:
                css: 'div.border-dotted:nth-of-type(2) span.pull-right'
                xpath: null
                multiple: true
                type: Text
            Exterior_color:
                css: 'div.border-dotted:nth-of-type(4) span.pull-right'
                xpath: null
                multiple: true
                type: Text
            Interior_color:
                css: 'div.border-dotted:nth-of-type(5) span.pull-right'
                xpath: null
                multiple: true
                type: Text
            title:
                css: 'a.btn-link'
                xpath: null
                multiple: true
                type: Text
            image:
                css: 'div.col-lg-3 img.img-responsive'
                xpath: null
                multiple: true
                type: Image
    
    '''
    
e = Extractor.from_yaml_string(yml)
data = e.extract(response.text)
print(url,"\n",data['cars_details'])
#page += 1
#try:
all_data += data['cars_details']
#except:
 #   break
        
```

```python
data = []
for item in all_data:
    
    my_dict = {}
    try:
        my_dict['mileage'] = item['mileage'][0]
        my_dict['Exterior_color'] = item['Exterior_color'][0]
        my_dict['Interior_color'] = item['Interior_color'][0]
        my_dict['title'] = item['title'][0]
        my_dict['image'] = item['image'][0]
    except:
        my_dict['mileage'] = None
        my_dict['Exterior_color'] = None
        my_dict['Interior_color'] = None
        my_dict['title'] = None
        my_dict['image'] = None
#     my_dict['Stock'] = None
#     my_dict['trim'] = None
    data.append(my_dict)
```
